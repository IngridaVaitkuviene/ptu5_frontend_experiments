```python
from bs4 import BeautifulSoup
import requests
```

```python
html = requests.get('http://delfi.lt')
html.status_code
```

```python
# pirmas variantas
if html.status_code == 200:
    delfi_soup = BeautifulSoup(html.content, "html.parser")
else:
    delfi_soup = html.status_code
delfi_soup
```

```python
# antras variantas

html = requests.get('http://delfi.lt').text
delfi_soup = BeautifulSoup(html, "html.parser")
delfi_soup
```

```python
antrastes = delfi_soup.select('.CBarticleTitle')
antrastes
```

```python
items = []
for item in antrastes:
    items.append(item.get_text())
items
```

```python
for item in items:
    if item.find(':') > -1:
        print(item)
```

```python
pirma_dalis = []
antra_dalis = []
for antraste in items:
    if ":" in antraste:
        padalinta_antraste = antraste.split(":")
        pirma_dalis.append(padalinta_antraste[0])
        antra_dalis.append(padalinta_antraste[1])
```

```python
pirma_dalis
```

```python
antra_dalis
```

```python
from random import shuffle
shuffle(antra_dalis)
```

```python
antra_dalis
```

```python
for dalys in range(len(pirma_dalis)):
    print(pirma_dalis[dalys], ':', antra_dalis[dalys])
```
